# Inverted Index

We will use spacy to tokenize, remove stopwords; punctuation and lemmatize

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

## The function to build an inverted index

```python
from collections import defaultdict

def build_inverted_index(documents: dict[str: str]):
    index = defaultdict(list)
    
    for doc_id, text in documents.items():
        doc = nlp(text)
        tokens = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
        
        for token in tokens:
            index[token].append(doc_id)
    
    return {term: sorted(docs) for term, docs in index.items()}
```

```python
docs = {
    "D1" : "Machine Learning is a branch of Artificial Intelligence.",
    "D2": "Data Mining deals with collecting, storing and extracting insights from the data.",
    "D3": "Machine Learning and Data Mining share a common interest i.e. both deal with data."
}
```

```python
index = build_inverted_index(docs)
index
```

## AND Query

```python
def and_query(index, terms):
    postings = [set(index.get(term, [])) for term in terms]
    postings.sort(key=len)
    result = postings[0]
    
    for p in postings[1:]:
        result = result.intersection(p)
    return result

and_query(index, ["machine", "data"])
```

## OR Query

```python
def or_query(index, terms):
    postings = [set(index.get(term, [])) for term in terms]
    postings.sort(key=len)
    result = postings[0]
    
    for p in postings[1:]:
        result = result.union(p)
    return result

or_query(index, ["machine", "data"])
```
