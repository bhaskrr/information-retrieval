# Stemming and Lemmatization

```python
text = '"Joy of Kosher" magazine fills a much-needed niche for kosher recipes, cookbook reviews, and the kosher lifestyle not found in other mainstream magazines.'
```

## Using NLTK

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text.lower())

# Perform tokenization and remove stopwords
filtered_tokens = [token for token in tokens if token not in stop_words]
filtered_tokens
```

```python
from nltk.stem import SnowballStemmer, WordNetLemmatizer

ss = SnowballStemmer(language="english")
lemmatizer = WordNetLemmatizer()

# Stemming
stems = [ss.stem(token) for token in filtered_tokens]
stems
```

```python
# Lemmatization
lemmas = [lemmatizer.lemmatize(token) for token in filtered_tokens]
lemmas
```
