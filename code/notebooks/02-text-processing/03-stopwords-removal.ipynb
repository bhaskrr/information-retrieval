# Stopwords Removal in Python

## Using nltk

```python
import nltk

nltk.download("stopwords")

from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words("english"))
```

```python
def remove_stopwords(tokens: list[str]) -> list[str]:
    return [token for token in tokens if token not in STOPWORDS]

tokens = ["the", "cat", "sat", "on", "the", "mat"]
print(remove_stopwords(tokens))
```

```python
# Building a custom stoplist:
# Extend the default list with domain-specific stopwords
custom_stopwords = STOPWORDS | {"http", "www", "com", "said", "also"}

def remove_stopwords_custom(tokens: list[str]) -> list[str]:
    return [token for token in tokens if token not in custom_stopwords]

tokens = ["he", "also", "works", "in", "www", "google", "com"]
print(remove_stopwords_custom(tokens))
```

## Using SpaCy

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
text = '"Joy of Kosher" magazine fills a much-needed niche for kosher recipes, cookbook reviews, and the kosher lifestyle not found in other mainstream magazines.'

doc = nlp(text)

# Print all the stop words in the the text
[token for token in doc if token.is_stop]
```

```python
# Print all the other tokens
[token for token in doc if not token.is_stop]
```
