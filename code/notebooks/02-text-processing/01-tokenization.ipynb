## Using nltk

To install nltk using uv

Run ```uv add nltk```

````python
from nltk.tokenize import word_tokenize
````

````python
# word_tokenize returns a tokenized copy of text, using NLTK’s
# recommended word tokenizer (currently an improved TreebankWordTokenizer along with PunktSentenceTokenizer for the specified language).

# Visit <https://www.nltk.org/api/nltk.tokenize.html> to know more.

tokens = word_tokenize(text=text, language="english")
tokens
````

## Using SpaCy

To install spacy, run `uv add spacy`

And download their english(small) model. To do that, we need to install `pip` too because spacy internally uses pip, run:

```bash
uv add pip
uv run spacy download en_core_web_sm
```

````python
import spacy

# spaCy’s trained pipelines can be installed as Python packages. This means that they’re a component of your application,
# just like any other module. They’re versioned and can be defined as a dependency in your requirements.txt.
# Trained pipelines can be installed from a download URL or a local directory, manually or via pip.
# Their data can be located anywhere on your file system.

# load spaCy's english model
nlp = spacy.load("en_core_web_sm")
````

````python
# When you call nlp on a text, spaCy will tokenize it and then call each component on the Doc, in order.
# It then returns the processed Doc that you can work with.

# A Doc is a sequence of Token objects.
# Visit <https://spacy.io/> to know more.

doc = nlp(text)

[token.text for token in doc]
````
